```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model


file_path = 'ss.csv.xlsx'
df = pd.read_excel(file_path)
print(df)
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area, df.price, color = 'red', marker = '*' )
```

```python
model = linear_model.LinearRegression()
model.fit(df[['area']], df.price)
```

```python
model.predict([[3300]])
```

```python
model.coef_  # coefficient = m ===> price = m * Area + b
```

```python
model.intercept_ # b
```

```python
135.78767123*3300 + 180616.43835616432   # reg.predict([[3300]])
```

```python
%matplotlib inline
plt.xlabel('area', fontsize = 20)
plt.ylabel('price', fontsize = 20)
plt.scatter(df.area, df.price, color = 'red', marker = '+')
plt.plot(df.area, model.predict(df[['area']]), color = 'blue')

# This result gives us LINEAR REGRESSION
```
