```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
iris.data
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
dfp = df.drop(['sepal length (cm)', 'sepal width (cm)'], axis = 'columns')
dfp.head()
```

```python
plt.scatter(df['petal length (cm)'], df['petal width (cm)'])
```

```python
km = KMeans(n_clusters = 2)
km
```

```python
y_predicted =  km.fit_predict(dfp[['petal length (cm)','petal width (cm)']])
y_predicted
```

```python
dfp['cluster'] = y_predicted
dfp.head()
```

```python
dfp.rename(columns={
    'petal length (cm)': 'petal_length_cm',
    'petal width (cm)': 'petal_width_cm'
}, inplace=True)
```

```python
dfp
```

```python
dfp[dfp.cluster==0]
```

```python
df1 = dfp[dfp.cluster==0]
df2 = dfp[dfp.cluster==1]
df1.petal_length_cm
```

```python
km.cluster_centers_
```

```python
df1 = dfp[dfp.cluster==0]
df2 = dfp[dfp.cluster==1]
plt.scatter(df1.petal_length_cm,df1['petal_width_cm'],color='green')
plt.scatter(df2.petal_length_cm,df2['petal_width_cm'],color='red')

plt.xlabel('petal_length_cm')
plt.ylabel('petal_width_cm')
plt.legend()
```

```python
scaler = MinMaxScaler()

scaler.fit(dfp[['petal_width_cm']])
dfp['petal_width_cm'] = scaler.transform(dfp[['petal_width_cm']])

scaler.fit(dfp[['petal_length_cm']])
dfp['petal_length_cm'] = scaler.transform(dfp[['petal_length_cm']])
dfp
```

```python
plt.scatter(dfp.petal_length_cm,dfp['petal_width_cm'])
```

```python
km = KMeans(n_clusters=2)
y_predicted = km.fit_predict(dfp[['petal_length_cm','petal_width_cm']]) 
y_predicted
```

```python
dfp['cluster']=y_predicted
dfp.head()
```

```python
km.cluster_centers_
```

```python
df1 = dfp[dfp.cluster==0]
df2 = dfp[dfp.cluster==1]
df3 = dfp[dfp.cluster==2]
plt.scatter(df1.petal_length_cm,df1['petal_width_cm'],color='green')
plt.scatter(df2.petal_length_cm,df2['petal_width_cm'],color='red')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid') 
plt.legend()
```

```python
k_rng = range(1, 10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(dfp[['petal_length_cm', 'petal_width_cm']])
    sse.append(km.inertia_) #inertia_ find the sse
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
