```python
import pandas as pd
df = pd.read_csv("heart.csv")
df
```

```python
import numpy as np
import scipy.stats as stats

numeric_cols = df.select_dtypes(include=[float, int]).columns # Select numaric variables

df_zscore = df.copy()
df_zscore[numeric_cols] = df[numeric_cols].apply(stats.zscore)

df_zscore.head()
```

```python
# Remove outlires
df_no_outliers = df[(df_zscore[numeric_cols] < 3).all(axis=1) & (df_zscore[numeric_cols] > -3).all(axis=1)]
df_no_outliers
```

```python

df1 = df_no_outliers.copy()
df1.Sex.replace(
    {
        'F': 0,
        'M': 1
    },
    inplace=True)

df1.ChestPainType.replace(
    {
        'ATA': 1,
        'NAP': 2,
        'ASY': 3,
        'TA':4
    },
    inplace=True
)

df1.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True)



df1.head()
```

```python
df2 = pd.get_dummies(df1, drop_first=True)
df2 = df2.replace({True: 1, False: 0})
df2.head()
```

```python

X = df2.drop("HeartDisease",axis='columns')
y = df2.HeartDisease

X.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=30)
```

```python
X_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()
model_rf.fit(X_train, y_train)
model_rf.score(X_test, y_test)
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python

from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()
model_rf.fit(X_train_pca, y_train)
model_rf.score(X_test_pca, y_test)
```
